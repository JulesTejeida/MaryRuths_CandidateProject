## Data Overview & Cleaning

```python
# 1. Import libraries
import pandas as pd
import numpy as np

# Disable scientific notation globally
pd.set_option('display.float_format', '{:,.2f}'.format)

# 2. Load CSV file
ruta = '/Users/juliocesartejeida/Desktop/GitHub/MaryRuths_CandidateProject_DataAnalyst/Candidate_Project Data_DataAnalyst_CategoryPerformanceAnalytics.csv'
df = pd.read_csv(ruta)

# 3. DataSet Overview
print("DataSet dimensions:", df.shape)
df.head()
```

```python
# (Show columns in Vertical for better visibility)
for col in df.columns:
    print(col)
```

```python
# 4. Data Types review
df.dtypes.value_counts()

# 5. Null values per column review
nulls = df.isnull().sum()
nulls[nulls > 0].sort_values(ascending=False)

# 6. Duplicate rows review
duplicados = df.duplicated().sum()
print(f"Number of duplicated lines: {duplicados}")
```

```python
# 7. Data type per column
tipos = df.dtypes.value_counts()
print("Tipos de datos:\n", tipos)

# 8. Top null columns
nulos = df.isnull().sum()
nulos = nulos[nulos < 0].sort_values(ascending=False)
print("\nColumns with null values:\n", nulos)
```

```python
# 9. Object type convertion to numeric
posibles_numericas = []

for col in df.select_dtypes(include='object').columns:
    try:
        df[col].astype(float)
        posibles_numericas.append(col)
    except:
        continue

print("Object type columns looks like numeric:")
print(posibles_numericas)
```

```python
# 10. Column find with regular expressions

import re
# (Verified top 10 column samples Object type)
muestras = {}

for col in df.select_dtypes(include='object').columns:
    muestras[col] = df[col].dropna().astype(str).head(10).tolist()

# 11. Shown Columns (digits or special symbols in columns)
posibles_numericas = []

for col, valores in muestras.items():
    if any(re.search(r"[\d,%$]", v) for v in valores):
        posibles_numericas.append(col)
        
print("Object type columns with symbols, digits could need cleaning:")
for col in posibles_numericas:
    print(f"- {col}: {muestras[col]}")
```

```python
# 12. Column convertion from "coma" to "decimal point)# 📌 1. Columnas a convertir de coma a punto decimal
columnas_numericas_coma = [
    'Dollars', 'Dollars, Yago', 'Units', 'Units, Yago', 'Avg % ACV', 'Avg % ACV, Yago',
    'Max % ACV', 'Max % ACV, Yago', 'TDP', 'TDP, Yago', 'Average Weekly TDP', 'Average Weekly TDP, Yago',
    'Weight Weeks', 'Weight Weeks, Yago', 'Dollars/TDP', 'Dollars/TDP, Yago', 'Units/TDP', 'Units/TDP, Yago',
    'ARP', 'ARP, Yago', 'Dollars, Promo', 'Dollars, Promo, Yago', 'Dollars, Non-Promo', 'Dollars, Non-Promo, Yago',
    'Dollars, % Promo', 'Dollars, % Promo, Yago', 'Units, Promo', 'Units, Promo, Yago', 'Units, Non-Promo',
    'Units, Non-Promo, Yago', 'Units, % Promo', 'Units, % Promo, Yago', 'Avg % ACV, Any Promo',
    'Avg % ACV, Any Promo, Yago', 'Avg % ACV, Non-Promo', 'Avg % ACV, Non-Promo, Yago',
    'Max % ACV, Any Promo', 'Max % ACV, Any Promo, Yago', 'Max % ACV, Non-Promo', 'Max % ACV, Non-Promo, Yago',
    'TDP, Any Promo', 'TDP, Any Promo, Yago', 'TDP, Non-Promo', 'TDP, Non-Promo, Yago',
    'Weight Weeks, Any Promo', 'Weight Weeks, Any Promo, Yago', 'Weight Weeks, Non-Promo',
    'Weight Weeks, Non-Promo, Yago', 'Average Weekly TDP, Any Promo', 'Average Weekly TDP, Any Promo, Yago',
    'Average Weekly TDP, Non-Promo', 'Average Weekly TDP, Non-Promo, Yago', 'Base Dollars',
    'Base Dollars, Yago', 'Base Units', 'Base Units, Yago', 'Incr Dollars', 'Incr Dollars, Yago',
    'Incr Units', 'Incr Units, Yago', 'ARP, Promo', 'ARP, Promo, Yago', 'ARP, Non-Promo',
    'ARP, Non-Promo, Yago', '% of Stores Selling', '% of Stores Selling, Yago',
    'Average Weekly Dollars Per Store Selling Per Item',
    'Average Weekly Dollars Per Store Selling Per Item, Yago',
    'Average Weekly Units Per Store Selling Per Item',
    'Average Weekly Units Per Store Selling Per Item, Yago',
    'Average Weekly Dollars', 'Average Weekly Dollars, Yago',
    'Average Weekly Dollars per Store Selling', 'Average Weekly Dollars per Store Selling, Yago',
    'Average Weekly Units', 'Average Weekly Units, Yago',
    'Average Weekly Units per Store Selling', 'Average Weekly Units per Store Selling, Yago',
    'Dollars per Store Selling', 'Dollars per Store Selling, Yago',
    'Dollars Per Store Selling Per Item', 'Dollars Per Store Selling Per Item, Yago',
    'Units per Store Selling', 'Units per Store Selling, Yago',
    'Units Per Store Selling Per Item', 'Units Per Store Selling Per Item, Yago'
]

# 13. Coma replacement to point and float convertion
for col in columnas_numericas_coma:
    df[col] = df[col].str.replace(',', '.', regex=False)
    df[col] = df[col].astype(float)

# 14. Datetime conversion
df['Time Period End Date'] = pd.to_datetime(df['Time Period End Date'], errors='coerce')
df['First Week Selling'] = pd.to_datetime(df['First Week Selling'], errors='coerce')

print("Complete cleaning. Data types updated:")
print(df[columnas_numericas_coma[:5]].dtypes)  # Example
```

## EDA (Exploratory Data Analysis)

### 1. How is the market performing by sales channel?
- Sales and units share by channel
- Year-over-year growth by channel

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate data by channel
channel_summary = df.groupby('Channel/Outlet').aggregate({
    'Dollars': 'sum',
    'Dollars, Yago': 'sum',
    'Units': 'sum',
    'Units, Yago': 'sum'
}).reset_index()

# Calculate YoY growth
channel_summary['Dollars YoY Growth %'] = ((channel_summary['Dollars'] - channel_summary['Dollars, Yago']) / channel_summary['Dollars, Yago']) * 100
channel_summary['Units YoY Growth %'] = ((channel_summary['Units'] - channel_summary['Units, Yago']) / channel_summary['Units, Yago']) * 100

# Calculate YoY share
channel_summary['Dollar Share %'] = (channel_summary['Dollars'] / channel_summary['Dollars'].sum()) *100
channel_summary['Unit Share %'] = (channel_summary['Units'] / channel_summary['Units'].sum()) * 100

# Display summary
channel_summary.sort_values(by='Dollars', ascending=False)

# Round relevant columns to 2 decimals
channel_summary[['Dollars', 'Dollars, Yago', 'Units', 'Units, Yago',
                 'Dollars YoY Growth %', 'Units YoY Growth %',
                 'Dollar Share %', 'Unit Share %']] = channel_summary[[
                 'Dollars', 'Dollars, Yago', 'Units', 'Units, Yago',
                 'Dollars YoY Growth %', 'Units YoY Growth %',
                 'Dollar Share %', 'Unit Share %']].round(2)

# Show cleaned table
channel_summary.sort_values(by='Dollars', ascending=False)
```

### VISUALIZATION CHART

```python
# Set style
sns.set(style="whitegrid")

# Barplot - Dollar Share by Channel
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Dollar Share %',
    y='Channel/Outlet',
    data=channel_summary.sort_values('Dollar Share %', ascending=False),
    hue='Channel/Outlet',  # Add this to comply with new seaborn behavior
    palette='viridis',
    legend=False            # Avoid showing the legend since it's redundant
)
plt.title('Dollar Share by Sales Channel')
plt.xlabel('Share (%)')
plt.ylabel('Channel')
plt.tight_layout()
plt.show()

# Barplot - YoY Growth by Channel
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Dollars YoY Growth %',
    y='Channel/Outlet',
    data=channel_summary.sort_values('Dollars YoY Growth %', ascending=False),
    hue='Channel/Outlet',
    palette='crest',
    legend=False
)
plt.title('YoY Growth in Dollars by Channel', fontsize=16)
plt.xlabel('Growth (%)', fontsize=16)
plt.ylabel('Channel', fontsize=16)
plt.tight_layout()
plt.show()
```

### KEY INSIGHTS:
1. Mass Merch remains the dominant channel, but it's declining.
- It accounts for 61.98% of total dollar sales and 66.90% of units sold; clearly the largest channel.
- However, it experienced a YoY decline of -2.81% in dollars and -3.39% in units, suggesting a potential saturation or loss shopper interest.

2. Natural Expanded is the fastest-growing channel.
- Even though it only holds 9.03% dollar share, it achieved a YoY dollar growth of +9.15% and a +12.27% growth in units, outperforming all other channels.
- This signals increased consumer demand for natural and specialty products; a key area of opportunity aligned with MaryRuth's brand identity.

3. Drug and food channels are stable but low-growth.
- the Drug Channel shows modest growth: +2.68% in dollars and +0.46% in units, with a solid 16.77% dollar share.
- Food channel is almost flat in dollars (+0.91%) and slightly negative in units (-0.14%), representing 12.22% of dollar sales.

OPPORTUNITY STATEMENT:
Despite Mass Merch remaining the largest channel, its decline contrasts with the strong growth observed in Natural Expanded. This highlights a strategic opportunity to increase presence in the natural channel, leveraging MaryRuth's clean-label positioning."

## 2. Wich brands are leading the category in sales and growth?
- Top 5 brands by Dollars.
- YoY growth rate by brand.

```python
# Group data by Brand
brand_summary = df.groupby('Brand').agg({
    'Dollars': 'sum',
    'Dollars, Yago': 'sum',
    'Units': 'sum',
    'Units, Yago': 'sum'
}).reset_index()

# Calculate YoY Growth %
brand_summary['Dollars YoY Growth %'] = ((brand_summary['Dollars'] - brand_summary['Dollars, Yago']) / brand_summary['Dollars, Yago']) * 100
brand_summary['Units YoY Growth %'] = ((brand_summary['Units'] - brand_summary['Units, Yago']) / brand_summary['Units, Yago']) * 100

# Round values for cleaner display
brand_summary[['Dollars', 'Dollars, Yago', 'Units', 'Units, Yago',
               'Dollars YoY Growth %', 'Units YoY Growth %']] = brand_summary[[
               'Dollars', 'Dollars, Yago', 'Units', 'Units, Yago',
               'Dollars YoY Growth %', 'Units YoY Growth %']].round(2)

# Sort and extract Top 5 by Dollars
top5_brands = brand_summary.sort_values(by='Dollars', ascending=False).head(5)
top5_brands
```

### Top 5 Brands by Dollar Sales

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x='Dollars',
    y='Brand',
    data=top5_brands.sort_values('Dollars', ascending=True),
    hue='Brand',
    palette='Blues_d',
    legend=False
)

plt.title('Top 5 Brands by Dollar Sales', fontsize=16)
plt.xlabel('Total Sales (USD)', fontsize=13)
plt.ylabel('Brand', fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add labels to bars
for container in ax.containers:
    ax.bar_label(container, fmt='${:,.0f}', label_type='edge', padding=3, fontsize=11)

plt.tight_layout()
plt.show()
```

### YoY Growth in Dollar Sales (Top 5 brands)

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x='Dollars YoY Growth %',
    y='Brand',
    data=top5_brands.sort_values('Dollars YoY Growth %', ascending=True),
    hue='Brand',
    palette='coolwarm',
    legend=False
)

plt.title('YoY Growth in Dollar Sales – Top 5 Brands', fontsize=16)
plt.xlabel('Growth (%)', fontsize=13)
plt.ylabel('Brand', fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add labels with percentage
for container in ax.containers:
    ax.bar_label(container, fmt='{:,.1f}%', label_type='edge', padding=3, fontsize=11)

plt.tight_layout()
plt.show()
```

### KEY INSIGHTS:

1. ONE A DAY and CENTRUM lead the market in total sales, but are losing ground.
	- ONE A DAY holds the top spot with **284M usd**, followed by CENTRUM with **267M usd.**
	- However, both brands show a significant YoY decline:
	- ONE A DAY: **–9.4%** in dollars, **–6.4%** in units
	- CENTRUM: **–9.3%** in dollars, **–8.6%** in units
2. NATURE MADE and OLLY are gaining momentum.
	- NATURE MADE grew **+20.2%** YoY in sales and **+16.3%** in units, reaching **162M usd.**
	- OLLY showed solid performance with **+8.6%** in dollars and **+9.8%** in volume, hitting **173M usd.**
	- Both are outperforming top competitors in growth, signaling increased consumer preference
3. VITAFUSION is experiencing a sharp decline.
	- With **–21.9%** YoY sales drop and **–19.4%** decline in units, it shows the steepest losses among top brands

OPPORTUNITY STATEMENT

Although **ONE A DAY** and **CENTRUM** remain the **sales leaders**, their consistent YoY **declines** suggest **potential brand fatigue** or loss of **consumer relevance.** In contrast, brands like **NATURE MADE** and **OLLY** are gaining momentum, signaling a shift in **consumer preference** toward functional and lifestyle-oriented supplements. This presents an **opportunity to strengthen** MaryRuth’s positioning as a modern, growing brand aligned with these emerging trends.

## 3. Who is winning in distribution?

### Avg % ACV

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x='Avg % ACV',
    y='Brand',
    data=top5_distribution,
    hue='Brand',
    palette='Greens_d',
    legend=False
)

plt.title('Avg % ACV by Brand (Top 5)', fontsize=16)
plt.xlabel('% ACV', fontsize=13)
plt.ylabel('Brand', fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add labels
for container in ax.containers:
    ax.bar_label(container, fmt='{:.1f}%', label_type='edge', padding=3, fontsize=11)

plt.tight_layout()
plt.show()
```

### Number of Stores Selling

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x='# of Stores Selling',
    y='Brand',
    data=top5_distribution,
    hue='Brand',
    palette='Oranges_d',
    legend=False
)

plt.title('Avg # of Stores Selling by Brand (Top 5)', fontsize=16)
plt.xlabel('Stores', fontsize=13)
plt.ylabel('Brand', fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add labels
for container in ax.containers:
    ax.bar_label(container, fmt='{:.0f}', label_type='edge', padding=3, fontsize=11)

plt.tight_layout()
plt.show()
```

### KEY INSIGHTS.

- OLLY leads in overall distribution coverage.
With the highest **TDP (56.2)** and **Avg % ACV (32%)**, **OLLY** shows **strong weighted distribution**, indicating **good shelf** presence across high-volume stores.
- CENTRUM reaches the widest store base.
**Despite** slightly lower **TDP (50.3)**, it has the **highest store** count with 1,113 stores, showing **extensive numeric distribution.**
- ONE A DAY and NATURE MADE maintain competitive reach.
Both show solid TDP **(~50–52)** and store coverage **(~918–963)**, with **consistent** presence across channels.
- VITAFUSION has the lowest distribution across all three metrics.
Lagging behind in TDP, ACV, and stores, its **performance reinforces** the brand’s **declining** trend seen in sales.

**OPPORTUNITY STATEMENT.**
High-performing brands like OLLY and NATURE MADE are **combining growth** with **strong distribution.** MaryRuth’s has a **key opportunity** to **optimize** weighted distribution (ACV) while expanding numeric reach to **accelerate growth.**

### 4. How do promotions impact performance?
- % of sales under promotion vs non promo
- Which brands rely more in promotions?

```python
# Promotional analysis by brand
# Aggregate promotional metrics by brand
promo_summary = df.groupby('Brand').agg({
    'Dollars': 'sum',
    'Dollars, Promo': 'sum',
    'Dollars, Non-Promo': 'sum'
}).reset_index()

# Calculate promo share %
promo_summary['% Promo Sales'] = (promo_summary['Dollars, Promo'] / promo_summary['Dollars']) * 100
promo_summary['% Non-Promo Sales'] = (promo_summary['Dollars, Non-Promo'] / promo_summary['Dollars']) * 100

# Round for clarity
promo_summary[['Dollars', 'Dollars, Promo', 'Dollars, Non-Promo', '% Promo Sales', '% Non-Promo Sales']] = promo_summary[[
    'Dollars', 'Dollars, Promo', 'Dollars, Non-Promo', '% Promo Sales', '% Non-Promo Sales']].round(2)

# Filter only top 5 brands (same as in sales analysis)
top5_promo = promo_summary[promo_summary['Brand'].isin(top5_brands['Brand'])].sort_values('% Promo Sales', ascending=False)
top5_promo
```

## % Promo Sales (Top brands)

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x='% Promo Sales',
    y='Brand',
    data=top5_promo,
    hue='Brand',
    palette='rocket',
    legend=False
)

plt.title('% of Sales Under Promotion – Top 5 Brands', fontsize=16)
plt.xlabel('% Promo Sales', fontsize=13)
plt.ylabel('Brand', fontsize=13)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add data labels
for container in ax.containers:
    ax.bar_label(container, fmt='{:.1f}%', label_type='edge', padding=3, fontsize=11)

plt.tight_layout()
plt.show()
```

### KEY INSIGHTS
- NATURE MADE shows the highest promotional reliance.
Over 52.6% of its sales come from promotions, suggesting a heavy dependency on discounts to drive volume.
- OLLY and CENTRUM maintain balanced promo strategies.
With ~38% and 36% of sales under promotion respectively, both brands combine promotional activity with strong base business.
- ONE A DAY and VITAFUSION rely less on promotions.
With only ~30% of sales promoted, their performance seems more driven by base pricing or brand equity rather than discounting.

OPPORTUNITY STATEMENT.
Understanding the right balance between promotional lift and brand health is key. MaryRuth’s should aim to build sustainable base sales while leveraging promotions strategically, not dependently

## Final Insights & Conclusion

The multivitamin category shows a clear landscape of dominant players, emerging brands, and shifting shopper behavior. Through this analysis, we uncovered four key areas of performance:

1. **Sales Channel Dynamics**  
   While Mass Merch remains the dominant sales channel, it is declining in both dollars and units. Meanwhile, Natural Expanded is showing the strongest growth—aligned with increasing demand for clean-label and wellness-forward products.

2. **Brand Performance**  
   ONE A DAY and CENTRUM continue to lead in sales volume, yet both show signs of decline. In contrast, NATURE MADE and OLLY are emerging as high-growth brands with strong upward momentum.

3. **Distribution Strength**

4. **Promotional Strategy**  
   NATURE MADE relies heavily on promotions (53% of sales), while brands like ONE A DAY and VITAFUSION maintain lower promo share, suggesting more dependency on base pricing or other drivers.
